```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error
```

```python
# Ubah kolom Tanggal Transaksi menjadi index jika belum berupa tipe datetime
```

```python
# Filter data untuk hanya mencakup kolom yang dibutuhkan (Quantity penjualan minyak goreng)
sales_data = data['Quantity']
```

```python
# Memisahkan data menjadi data latih dan data uji
train_size = int(len(sales_data) * 0.8)  # Menggunakan 80% data sebagai data latih
train_data, test_data = sales_data[:train_size], sales_data[train_size:]
```

```python
df_train = data.iloc[:-20]
df_test = data.iloc[-30:]
```

```python
# Visualisasi data latih
plt.figure(figsize=(10, 6))
plt.plot(df_train.index, df_train['Quantity'], label='Data Latih', color='blue')
plt.title('Data Latih Penjualan Minyak Goreng')
plt.xlabel('Tanggal Transaksi')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()

# Visualisasi data uji
plt.figure(figsize=(10, 6))
plt.plot(df_test.index, df_test['Quantity'], label='Data Uji', color='red')
plt.title('Data Uji Penjualan Minyak Goreng')
plt.xlabel('Tanggal Transaksi')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Membuat model Holt-Winters Exponential Smoothing dengan jenis komponen "additive"
model_add = ExponentialSmoothing(df_train['Quantity'], trend='add', seasonal='add', seasonal_periods=30).fit()

# Melakukan peramalan untuk data uji
forecast_add = model_add.forecast(len(df_test))

# Menampilkan hasil peramalan
plt.figure(figsize=(10, 6))
plt.plot(df_train.index, df_train['Quantity'], label='Data Latih', color='blue')
plt.plot(df_test.index, df_test['Quantity'], label='Data Asli', color='green')
plt.plot(df_test.index, forecast_add, label='Peramalan Additive', color='red')
plt.title('Peramalan Penjualan Minyak Goreng dengan Holt-Winters (Additive)')
plt.xlabel('Tanggal Transaksi')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Melakukan peramalan untuk data uji
forecast = model_add.forecast(len(test_data))

# Menghitung Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(test_data, forecast))

print(f'Hasil peramalan: {forecast}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
